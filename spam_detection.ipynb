```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('sum_data.csv')
data.head()
```

```python
data.shape
```

```python
data = data.dropna()
data.shape
```

```python
data['len'] = data['reviewText'].str.count(' ')
data.head()
```

```python
df = data.drop(data[data.len < 15].index)
df.head()
```

```python
df_groupby_id = df.groupby(['reviewText']).size().to_frame('count')
# df_groupby_id[df_groupby_id['count'] >1].sort_values('count' ,ascending=False)
df = pd.merge(df, df_groupby_id, on=['reviewText'])
df
```

```python
df = df[df['count'] > 1]
df = df.groupby(['reviewText', 'reviewerID',  'productID']).sum()
df = df.reset_index()
df
# df['reviwerID'].unique().shape
```
