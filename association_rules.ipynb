```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from apyori import apriori
%matplotlib inline
```

```python
data = pd.read_csv('sum_data.csv')
data.head()
```

```python
df_groupby_id = data.groupby(data['reviwerID']).size().to_frame('count_reviewerID')
df_count = df_groupby_id.sort_values(by=['count_reviewerID'], ascending = False)[0:1000]
# df_count = df_count.set_index('index')
df_count = df_count.reset_index()
# df_count = df_count.reset_index(drop = True)
df_count.head()
```

```python
top1000 = pd.DataFrame(columns=df_count['reviwerID'])
for index, row in df_count.iterrows():
    tmp = data[data.reviwerID == row.reviwerID].reset_index()
    tmp = tmp['productID']
    top1000[row.reviwerID] = tmp
    
#     top1000 = top1000.append(tmp[['reviwerID','productID']])
# top1000 = top1000.reset_index(drop = True)
# top1000[136670] = data[data.reviwerID == 136670]['productID']
top1000 = top1000.T
top1000 = top1000.fillna(0)
top1000.head()
```

```python
top1000.to_csv('top1000_reviewer.csv')
print("save top1000_reviewer.csv complete")
```

```python
top1000 = pd.read_csv('top1000_reviewer.csv', header = None)
top1000 = top1000.drop(0).reset_index(drop = True)
top1000 = top1000.drop(columns=[0])
top1000.head()
```

```python
records = []  
for i in range(0, 1000):  
    tmp = []
    for j in range(0,431):
        if(top1000.values[i,j] != '0'):
            tmp.append(top1000.values[i,j])
        else:
            break
    records.append(tmp)
#     records.append([str(top1000.values[i,j]) for j in range(0, 431)])
records[1]
```

```python
association_rules = apriori(records, min_support=0.02, min_confidence=0.5, min_length=2)  
association_results = list(association_rules)  
```
