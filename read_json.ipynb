```python
import sys
import json
import pandas as pd
```

```python
tweets_data_path = 'amz_review.json'

tweets_data = []
tweets_file = open(tweets_data_path, "r")

count = 0
for line in tweets_file:
    try:
        if (line == '\n'):      # skip empty lines
            continue
        count = count + 1
        tweet = json.loads(line)
        tweets_data.append(tweet)
        if count%10000 == 0:
            sys.stdout.write('.')
        if count%700000 == 0:
            sys.stdout.write('\n')
    except Exception as e:
        print(e)
        continue
print(f"\n{count} tweets read.")    # 1689188
tweets = pd.DataFrame()
tweets['reviwerID'] = list(map(lambda tweet: tweet.get('reviewerID', None), tweets_data))
tweets['productID'] = list(map(lambda tweet: tweet.get('asin', None), tweets_data))
tweets['reviewText'] = list(map(lambda tweet: tweet.get('reviewText', None), tweets_data))
tweets['rating'] = list(map(lambda tweet: tweet.get('overall', None), tweets_data))
tweets['summary'] = list(map(lambda tweet: tweet.get('summary', None), tweets_data))
tweets['helpful'] = list(map(lambda tweet: tweet.get('helpful', None), tweets_data))
tweets['unixTime'] = list(map(lambda tweet: tweet.get('unixReviewTime', None), tweets_data))
tweets['reviewTime'] = list(map(lambda tweet: tweet.get('reviewTime', None), tweets_data))


# tweets

# tweets.to_csv('data.csv',index=False)
# print("save csv completed")
```
