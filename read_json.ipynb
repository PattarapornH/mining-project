```python
import sys
import json
import pandas as pd
```

```python
amz_data_path = 'amz_review.json'

amz_data = []
amz_review_file = open(amz_data_path, "r")

count = 0
for line in amz_review_file:
    try:
        if (line == '\n'):      # skip empty lines
            continue
        count = count + 1
        amz = json.loads(line)
        amz_data.append(amz)
        if count%10000 == 0:
            sys.stdout.write('.')
        if count%700000 == 0:
            sys.stdout.write('\n')
    except Exception as e:
        print(e)
        continue
print(f"\n{count} amz_review read.")    # 1689188
amz_review = pd.DataFrame()
amz_review['reviwerID'] = list(map(lambda amz: amz.get('reviewerID', None), amz_data))
amz_review['productID'] = list(map(lambda amz: amz.get('asin', None), amz_data))
amz_review['reviewText'] = list(map(lambda amz: amz.get('reviewText', None), amz_data))
amz_review['helpful'] = list(map(lambda amz: amz.get('helpful', None), amz_data))
```

```python
amz_review.head()
```

```python
tmp = pd.DataFrame()
tmp['help'] = amz_review["helpful"]
tmp['help'] = tmp['help'].map(lambda x: ''.join(str(x)).lstrip('\[').rstrip('\]'))
tmp.head()
```

```python
split_helpful = tmp["help"].str.split(",", n = 1, expand = True)
split_helpful
amz_review['helped'] = split_helpful[0].astype(int)
amz_review['sum_vote'] = split_helpful[1].astype(int)
# amz_review.head()
```

```python
amz_review['unhelped'] = amz_review['sum_vote'] - amz_review['helped']
# data[150:200]
```

```python
amz_review['sum_helped'] = amz_review['helped'] - amz_review['unhelped']
# data = data.drop(data['sum_vote'])
amz_review.head()
# data[150:200]
```

```python
amz_review = amz_review.drop(columns=['sum_vote'])
amz_review.head()
```

```python
amz_review.to_csv('sum_data.csv', index_label="index")
print("save sum_data.csv completed")
```
